```python
import boto3
from torch.utils.data import Dataset
from io import BytesIO
from torchvision import transforms

AWS_BUCKET_NAME = 'deprojteam07-labeledrawdata'
#가져올 데이터 버전
#prefix = 'split_1'


class S3ImageDataset(Dataset):
    def __init__(self, bucket_name, transform=None):
        self.s3 = boto3.client('s3')
        self.bucket_name = bucket_name

        self.image_keys = self.get_image_keys()

    def get_image_keys(self):
        image_keys = []
        obj_list = self.s3.list_objects(Bucket=self.bucket_name)
        for content in obj_list['Contents']:
            image_keys.append(content['Key'])
        return image_keys
    
    def __len__(self):
        return len(self.image_keys)
    
    def __getitem__(self, idx):
        image_key = self.image_keys[idx]
        response = self.s3.get_object(Bucket=self.bucket_name, Key=image_key)
        image_data = response['Body'].read()
        image = Image.open(BytesIO(image_data))

        if self.transform:
            image = self.transform(image)

        return image
    

transform = transforms.Compose([
                                transforms.Resize(size = (64,64)),
                                transforms.ToTensor()
                                ])

dataset = S3ImageDataset(bucket_name=AWS_BUCKET_NAME, transform = transform)

print(dataset)
```
